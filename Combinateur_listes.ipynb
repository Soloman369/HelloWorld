```python
from itertools import product
noms = ['Hamza' , 'Othman' , 'Soueiman']
adjs = ['Sgue3' , 'Tigog' , 'Nze9']
```

```python
def combinateur_de_liste(noms, adjs):
    
    liste_finale1 = []
    
    for i in noms:
        for j in adjs:
            liste_finale1.extend((i , j))
    return (liste_finale1)

liste_finale = combinateur_de_liste(noms,adjs)
print(combinateur_de_liste(noms,adjs))
```
