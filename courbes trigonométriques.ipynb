```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y = []
z = []
```

```python
for i in np.linspace(-10, 10, 100):
    x.append(i)
    y.append(math.cos(i))
    z.append(math.sin(i))
```

```python
plt.cla()
plt.title('Courbes trigonomÃ©triques ðŸ˜…')
plt.plot(x, y, label='Cos')
plt.plot(x, z, label='Sin')
plt.gca().legend(loc = 'upper left')
plt.show()
```
